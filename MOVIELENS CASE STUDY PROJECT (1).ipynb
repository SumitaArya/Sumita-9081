```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model.intercept_
```

```python
log_model.coef_
```

```python
y_pred=log_model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
log_model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```
